## Basic Computer Vision with ML

Fashion MNIST is a dataset with 70,000 images in 10 different categories, each category has 7000 images - each image is a 28x28 image of a Fashion item - shirts, pants, shoes etc.

Fashion MNIST dataset is available in **keras** - keras.datasets.fashion_mnist module.  
Its data can be accessed using load_data() method. Data is stored in the format : (img_train_data, cat_train_data), (img_test_data, cat_test_data).  
Here
* img_train_data - training data images
* cat_train_data - training data category
* img_test_data - test data images
* cat_test_data - test data category

Categories are numbers [0-9], while images are (28,28) shape numpy arrays. Categories are numbers in most datasets because :
* Computers work faster with numbers than with strings
* We can then assign appropriate labels to the numbers based on language or context.


```python
"""
Retreive data
"""
import tensorflow
import keras.datasets

fashion = keras.datasets.fashion_mnist
(img_train_data, cat_train_data),(img_test_data, cat_test_data) = fashion.load_data()
print(f"Data Container Shape :\nTraining Image Data = {img_train_data.shape}\nTraining Categorical Data = {cat_train_data.shape}\nTest Image Data = {img_test_data.shape}\nTest Image Data = {cat_test_data.shape}\n")
```

```python
"""
Display random index image data and categorical data
"""
import random

index = random.randint(0,img_train_data.shape[0])
print(f"Data at index {index} is : \n{img_train_data[index]}\n")
print(f"Category at index {index} is : {cat_train_data[index]}")
```

Since image data is in the range [0,255], we need to normalize this data to get better results.

```python
img_train_data_norm = img_train_data/255.0
img_test_data_norm = img_test_data/255.0
```

Now we construct a Neural Network which will take in a 28x28 image and go through all the pixel values in the image and try to determine the relationship between the pixel values and the category assigned to it.

The Neural Network we use is a **Sequential** Neural Network with layers :
  * **Flatten** - Flattens a n-dimensional numpy array
  * **Dense** - It performs the mathematical operation output=activation(dot(input,kernel)+bias),
  computing a weighted sum of all inputs, adding a bias vector, and applying an activation function.

We use Flatten layer (keras.layers.Flatten()) to change (n,n) image to (n*n,) array.  
We use Dense layer (keras.layers.Dense()) with 128 units and activation function "tensorflow.nn.relu"
- units is adjusted based on the performance of the model, 128 is a good starting point.
- "tensorflow.nn.relu" - computes rectified linear  
This will create 128 functions each with different weights. The weights are adjusted in each run such that the combination of the functions together will to obtain the output category accordingly.

We use Dense layer again on exit with 10 units and activation function "tensorflow.nn.softmax"
- this is because we require 10 output values one for each category hence the units parameter is set to 10.
- "tensorflow.nn.softmax" - computes softmax activation

```python
#define model
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation=tensorflow.nn.relu),
    keras.layers.Dense(10, activation=tensorflow.nn.softmax)
    ])
```

After we create and define our model, we compile the model using model_name.compile() method. This method takes in "optimizer" and "loss" as parameters.

**loss** determines the method of calculation of loss. It gives us a measure of how good or how bad the results are. Here we are using 'sparse_categorical_crossentropy'.

**optimizer** determines the optimization function we use to generate new parameters for the model's neurons which are used to get better results. Here we use "tensorflow.keras.optimizers.Adam()" ( Deprecated - 'tensorflow.train.AdamOptimizer()').

```python
#compile model
model.compile(optimizer=tensorflow.keras.optimizers.Adam(), loss='sparse_categorical_crossentropy')
```

### Activation functions
**tensorflow.nn.relu** - Rectified Linear Unit - This filters out negative numbers from the group.

**tensorflow.nn.softmax** - Softmax activation - This activation function returns the probability of the data belonging to each of the categories.

### Training

After compilation, we train the model using the training data. This is done using "model_name.fit(train_data, train_category, epochs)"

* **train_data** - data on which training model will work on
* **train_category** - categories into which data is actually classified
* **epochs** - number of iterations we want out model to iterate over for improvement of the model.


```python
model.fit(img_train_data_norm, cat_train_data, epochs=5)
```

### Using model on test data
We can use this model on test data to verify the categories the model will categorize the test data into with the actual categories they belong to. We do this using the **"model.evaluate(test_data, test_categories)"** method. This gives us **"test_acc"** as output.

Here
* test_data - data used to test the model
* test_categories - actual categories for test_data
* test_acc - Accuracy of the model on test_data

```python
#print(type(img_test_data), type(cat_test_data))
test_loss = model.evaluate(img_test_data, cat_test_data)
print(f"Test Loss for Model = {test_loss}")
```

Now we can use an image on our Machine Learning model to get an estimate of what fashion object our model is looking at. This can be done using the **"model.predict(images)"** method.

Here **images** is a (a, n, m) shaped numpy array containing multiple images.

**model.predict()** returns the list of list of probabilities of the image belonging to the category for each category for each image. Format :  
[  
  [image1_category1_probability, image1_category2_probability, ...,image1_categoryn_probability],  
[image2_category1_probability, image2_category2_probability, ...,image2_categoryn_probability],  
...  
[imageA_category1_probability, imageA_category2_probability, ...,imageA_categoryn_probability]
  
]

```python
import numpy

#Using the randomly generated index
index = random.randint(0,img_train_data.shape[0])

image = img_train_data[index]
category = cat_train_data[index]

image_norm = image/255.0
images = numpy.array([image_norm])
print(type(image_norm), images.shape)
probabilities = list(model.predict(images)[0])
print(probabilities)
```

```python
predict_cat = probabilities.index(max(probabilities))
print(f"Predicted Category = {predict_cat}\nActual Category = {category}")
```
