## Introducing Convolutional Neural Networks

In the previous chapter we used Neural Network to detect what category of fashion wear an object in an image belonged to. This performed well but there are issues with the model.

The object in the image needs to be centered and the object must be the only item in the image (empty background). In the real world this is not the case as multiple objects may be in an image and the objects may not be centered in the image.

For situations like this, we use **Convolutional Neural Networks (CNNs)**. In this approach we first filter the images before we train the deep neural network. After filtering the images, image features come in the forefront - which are spotted and then categorized using a Deep Neural Network.

**Filters** are matrices which are multiplied to the image to obtain only a certain feature of the image that we want. Example filters - horizontal edge filters, vertical edge filters, Gaussian Blur filters.

**Pooling** operation groups the pixels in an image and filters them down into a subset. For example we can take an image and perform **2x2 max pooling** which takes the highest value of a pixel in the 2x2 section of the image and places it in the output image. This results in an image with half the resolution.

Filters are learned by the CNNs. When we pass an image into the Convolutional Layer, the layer will initialize a set of filters with random parameters and pass over the image. These filtered images give us the features of an image depending on the filter used (Edge filters give out edges, etc).The results of these are fed into the next layer and matching is performed by the Neural Network. Over time, filters which give the image outputs best matching the categories will be learned and this process is called **Feature Extraction**.

Lets model a CNN for the MNIST fashion accessories categorisation :

```python
"""
Retreive data
"""
import tensorflow
import keras.datasets

fashion = keras.datasets.fashion_mnist
(img_train_data, cat_train_data),(img_test_data, cat_test_data) = fashion.load_data()
print(f"Data Container Shape :\nTraining Image Data = {img_train_data.shape}\nTraining Categorical Data = {cat_train_data.shape}\
\nTest Image Data = {img_test_data.shape}\nTest Image Data = {cat_test_data.shape}\n")
```

```python
#Normalize the Images
img_train_data_norm = img_train_data / 255.0
img_test_data_norm = img_test_data / 255.0
```

```python
#Define CNN
model = keras.models.Sequential([
    keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(28, 28, 1)),
    keras.layers.MaxPooling2D(2, 2),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

Here we are using the following layers :
* **keras.layers.Conv2D()** - This layer is a 2d Convolition layer - creates a convolution kernel convolved with the image in the spatial dimensions to produce a tensor of outputs. It is mainly used to detect local patterns like edges, textures and shapes in grid like data. Key parameters are :
  * filters - 64 - number of output feature maps
  * kernel_size - (3,3) - kernel size of the convolution filter
  * strides - default (1,1) - step size of convolution
  * padding - 'valid' or 'same' - to determine whether input is padded to preserve spatial dimensions
  * activation - 'relu' - activation function
  * input_shape - (28,28,1) - shape of input
* **keras.layers.MaxPooling2D()** - This layer is a 2d MaxPooling layer - Reduce dimensionality by taking the maximum value over a defined window (**pool_size**), shifting by a specified **strides** value(default value - pool_size), and applying **padding** (either "valid" or "same") to control output size.

We can add multiple groups of Conv2D and MaxPooling2D layers to break down the image and try to learn from lots of abstract features.

Using this stack of Conv2D and MaxPooling2D layers the model can categorize images better using the features in the image instead of using just the pixel values.

```python
#compile model
model.compile(optimizer=tensorflow.keras.optimizers.Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#Use model on training data
model.fit(img_train_data_norm, cat_train_data, epochs=5)
```

```python
#Evaluate model
test_loss, test_acc = model.evaluate(img_test_data_norm, cat_test_data)
print(f"Test Loss = {test_loss}\nTest Accuracy = {test_acc}")
```
