## Introduction to Machine Learning

Consider equation
y = 2x-1

Lets construct two lists with values following this equation.

```python
import random

start, end = -20, 20
size = 30

random.seed(10)

x_arr = [random.randint(start,end) for _ in range(size)]
y_arr = [2*x-1 for x in x_arr]

#x_arr = [-1,0,1,2,3,4]
#y_arr = [2*x-1 for x in x_arr]

#Normalise data
x_arr_norm = [x/20 for x in x_arr]
y_arr_norm = [y/41 if y<0 else y/39 for y in y_arr]
```

```python
print(f"Values of 'x' : {x_arr}")
print(f"Values of 'y' : {y_arr}")
```

Using Keras **Sequential** model we can estimate the pattern which matches each number to the other.

```python
import keras
import numpy

#Sequential model
seq_model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

#Compile the model
seq_model.compile(optimizer='sgd', loss='mean_squared_error')

#Convert data into numpy arrays
xs = numpy.array(x_arr_norm, dtype=float)
ys = numpy.array(y_arr_norm, dtype=float)
```

For Sequential model, we are using **Dense** Layer which has one neuron (units=1) and data of shape [1].

We provide compiler with an 'optimizer' and a 'loss' parameter. Machine learning models make a guess as to what the function is which yields the output value we see. The model calculates how good or how bad a guess is using the **loss** function. It then uses the **optimiser** function to generate another guess.

The combination of the **loss** and the **optimiser** function will get us closer and closer to the actual function.

# Model fitting
**Model fitting** is the process of adjusting a machine learning model's internal parameters (such as weights and biases) to minimize the difference between its predictions and the actual target data, typically using optimization techniques.

**model.fit()** function is the primary method used to train a model by iterating over the dataset for a specified number of **epochs**, calculating **loss**, and updating **weights**.

The fitting process generally involves forward passes to generate predictions, loss calculation to measure error, and backward passes (backpropagation) to compute gradients for weight updates

```python
num_epochs = 500
seq_model.fit(xs, ys, epochs=num_epochs)
```

**model.predict()** is a method used to generate predictions for new or unseen data after a model has been trained. It only needs the input data and returns the model's outputs, such as probabilities, class labels, or continuous values.

The method accepts input data in the form of NumPy arrays or TensorFlow datasets and processes them in batches, automatically handling inference mode (e.g., deactivating dropout layers). The structure of the output depends on the model's final layer.

```python
test_arr = [10, 2, -3, 0]
#When training data is normalised - test data needs to be normalised
norm_test_arr = [i/20 for i in test_arr]

results = seq_model.predict(numpy.array(norm_test_arr, dtype=float))
#Denormalize the output
results = [i*39 if i>0 else i*41 for i in results]
#results = [i*40 for i in results]
results
```
